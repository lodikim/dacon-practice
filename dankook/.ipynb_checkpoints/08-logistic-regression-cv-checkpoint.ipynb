# 08-Logistic Regression CV

### Imports

```python
import lightgbm as lgb
from matplotlib import pyplot as plt
from matplotlib import rcParams
import numpy as np
from pathlib import Path
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
import seaborn as sns
```

```python
rcParams['figure.figsize'] = (16, 8)
plt.style.use('fivethirtyeight')
pd.set_option('max_columns', 100)
pd.set_option("display.precision", 4)
```

### Load Train Data

```python
data_dir = Path('../../dacon-inputs/dankook/input')
sub_dir = Path('../../dacon-inputs/dankook/sub')
feature_dir = Path('../../dacon-inputs/dankook/feature')
val_dir = Path('../../dacon-inputs/dankook/val')
tst_dir = Path('../../dacon-inputs/dankook/tst')

trn_file = data_dir / 'train.csv'
tst_file = data_dir / 'test.csv'
sample_file = data_dir / 'sample_submission.csv'

target_col = 'class'
n_fold = 5
n_class = 3 # b/c class is 0, 1, 2?
seed = 42
```

```python
data_dir.resolve()
```

```python
algo_name = 'lrcv'
feature_name = 'polyfeature'
model_name = f'{algo_name}_{feature_name}'

feature_file = feature_dir / f'{feature_name}.csv'
p_val_file = val_dir / f'{model_name}.val.csv'
p_tst_file = tst_dir / f'{model_name}.tst.csv'
sub_file = sub_dir / f'{model_name}.csv'
```

### Make Polynomial Feature

```python
df = pd.read_csv(feature_dir / 'feature.csv', index_col=0)
print(df.shape)
df.head()
```

```python
scaler = StandardScaler()
poly = PolynomialFeatures(2)
X = poly.fit_transform(scaler.fit_transform(df.drop(target_col, axis=1)))
feature_names = poly.get_feature_names(df.columns)
feature_names
```

```python
df_poly = pd.DataFrame(data=X, columns=feature_names, index=df.index)
df_poly[target_col] = df[target_col]
df_poly.head()
df_poly.to_csv(feature_file)
```

```python
del df_poly, df
```

```python
df = pd.read_csv(feature_file, index_col=0)
print(df.shape)
df.head()
```

```python
y = df[target_col].values[:320000]
df.drop(target_col, axis=1, inplace=True)
trn = df.iloc[:320000].values
tst = df.iloc[320000:].values
feature_name = df.columns.tolist()
print(y.shape, trn.shape, tst.shape)
```

### Stratified K-Fold Cross Validation

```python
cv = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)
```

### Train Logistic Regression Classifier

```python
p_val = np.zeros((trn.shape[0], n_class))
p_tst = np.zeros((tst.shape[0], n_class))
for i, (i_trn, i_val) in enumerate(cv.split(trn, y), 1):
    print(f'training model for CV #{i}')
    clf = LogisticRegression(multi_class='multinomial')
    clf.fit(trn[i_trn], y[i_trn])
    p_val[i_val, :] = clf.predict_proba(trn[i_val])
    p_tst += clf.predict_proba(tst) / n_fold
```

```python
print(f'{accuracy_score(y, np.argmax(p_val, axis=1)) * 100:.4f}%')
```

```python
print(p_val.shape, p_tst.shape)
```

```python
# Stacking에 사용
np.savetxt(p_val_file, p_val, fmt='%.6f', delimiter=',')
np.savetxt(p_tst_file, p_tst, fmt='%.6f', delimiter=',')
```

### Visualize Feature Importance

```python
clf.coef_.shape
```

```python
imp = pd.DataFrame({'feature': df.columns, 'importance': clf.coef_[1, :].T})
imp = imp.sort_values('importance').set_index('feature')
imp.plot(kind='barh', figsize=(8, 32))
```

### Create Submission File

```python
sub = pd.read_csv(sample_file, index_col=0)
print(sub.shape)
sub.head()
```

```python
sub[target_col] = np.argmax(p_tst, axis=1)
sub.head()
```

```python
sub[target_col].value_counts()
```

### Saving Submission File

```python
sub.to_csv(sub_file)
```
